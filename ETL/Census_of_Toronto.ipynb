# This ETL will extract data from a Wikipedia page to import into DB. 
## This a dataset of census from 2011 to 2016

```python
#import Libraries
import pandas as pd
from sqlalchemy import create_engine,inspect
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

```python
url = 'https://en.wikipedia.org/wiki/Demographics_of_Toronto#cite_note-2011censuspop-18'
```

```python
tables = pd.read_html(url)
```

```python
tables[0]
```

```python
df_census_toronto= tables[0]
```

```python
df_census_toronto.columns = ['year', 'city','urban','cma','gta','GTHA', 'a','b']
```

```python
df_table = df_census_toronto.iloc[2:]
```

```python
df_table_clean = df_table[['year','city','gta']]
```

```python
df_table_clean = df_table_clean[df_table_clean['year']>'2000']
```

```python
df_table_clean
```

```python
# open connection with the local DB (postgres)
rds_connection_string = "postgres:filho21@localhost:5432/Crimes"
engine = create_engine(f'postgresql://{rds_connection_string}')

# automap_base to map all the columns into object
Base = automap_base()
Base.prepare(engine, reflect=True)
session = Session(engine)

City = Base.classes.city
Census = Base.classes.census


```

```python
# insert data into census table
engine.execute("delete from census")
session.commit()
for index, row in df_table_clean.iterrows():
    city_population =  str(row['city'])[0:9]
    gta_population = str(row['gta'])[0:9]
    year = row['year']
    city_population = float(city_population.replace(',',''))
    id_city = session.query(City.id).filter(City.name=='Toronto').first()
    session.add(Census(id_city = id_city, year=year,population = float(city_population)))
    session.commit()
```
